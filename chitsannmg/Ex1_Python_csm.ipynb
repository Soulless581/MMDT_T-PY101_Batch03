# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
seconds = int(input("Enter seconds :"))
minutes = int(seconds)//60
left_seconds = int(seconds) - (int(minutes)*60)
print("{} minutes and {} seconds".format(minutes,left_seconds))
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math
value = float(input("Enter a number :"))
sine_value = math.sin(value)
cosine_value = math.cos(value)
tangent_value = math.tan(value)
print( "sine value {} , cosine_value {} , tangent_value {}".format(sine_value,cosine_value,tangent_value))
```
